```python
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
#Qual a receita total?

df["Valor Venda"].sum()
```

```python
#Qual o custo total
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"])
```

```python
df.head()
```

```python
#custo total
round(df["Custo"].sum(), 2)
```

```python
#criando uma coluna de lucro
#calculando o lucro, valor da venda - custo

df["Lucro"] = df["Valor Venda"] - df["Custo"]
```

```python
df.head()
```

```python
#lucro total
round(df["Lucro"].sum(), 2)
```

```python
#Criando uma coluna de tempo de envio
df["Tempo_envio"] = df["Data Venda"] - df["Data Envio"]
```

```python
df.head()
```

**agora queremos saber a médio de tempo de envio para cada marca, e pra isso precisamos transformar a coluna Tempo_envio em númerica**

```python
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head()
```

```python
#média de tempo de envio por marca
df.groupby("Marca")["Tempo_envio"].mean()
```

```python
#Verificando se temos valores faltantes
df.isnull().sum()
```

```python
#Vamos agrupar por ano e por marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
#melhorando a visualização dos lucros
pd.options.display.float_format = "{:20,.2f}".format
```

```python
#Resetando o index ()
Lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
Lucro_ano
```

```python
#qual o total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
#Grafico do total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total de produtos vendidos")
plt.xlabel("Total")
plt.ylabel("Produtos");
```

```python
#Total de lucro por ano(bar na vertical e barh na horizontal)
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().sort_values(ascending=True).plot.barh(title="Lucro por ano");
plt.xlabel("Lucro")
plt.ylabel("Ano");
```

```python
#Selecionando as vendas só de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
df_2009.head()
```

```python
#lucro de 2009 por mês
df_2009.groupby(df_2009['Data Venda'].dt.month)["Lucro"].sum().plot(title="lucro x mês")
plt.xlabel("Mês")
plt.ylabel("Lucro");
```

```python
#lucro por marca (xticks define as os nomes das colunas de forma horizontal x vertical)
df_2009.groupby("Marca")["Lucro"].sum().sort_values(ascending=False).plot.bar(title="Marca x Lucro", color="green")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation="horizontal")
```

```python
df_2009.groupby("Classe")["Lucro"].sum().sort_values(ascending=False).plot.bar(title="lucro x Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation="horizontal");
```
